# 2D Line plot
- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
x = [1,2,3,4,5,6,7,8]
y = [10,22,5,47,89,63,56, 40]
plt.plot(x , y)
plt.show()
```

```python
# from a pandas dataframe (use sharma-kohli dataset )
df = pd.read_csv('./Datasets/sharma-kohli.csv')
df.head()
```

```python
plt.plot(df['index'], df['RG Sharma'] )
```

```python
plt.plot(df['index'], df['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])
```

```python
# labels and title
plt.plot(df['index'], df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(df['index'], df['V Kohli'], c="blue", linewidth=2, linestyle="solid", marker="o", markerfacecolor="red")
plt.plot(df['index'], df['RG Sharma'], c = "red", linewidth=2, marker="o")
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
```

```python
# legend -> location
plt.plot(df['index'], df['V Kohli'], c="blue", linewidth=2, linestyle="solid", marker="o", label="Virat Kohli")
plt.plot(df['index'], df['RG Sharma'], c = "red", linewidth=2, marker="o", label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
plt.legend()
```

```python
# limiting axes
plt.plot(df['index'], df['V Kohli'], c="blue", linewidth=2, linestyle="solid", marker="o", label="Virat Kohli")
plt.plot(df['index'], df['RG Sharma'], c = "red", linewidth=2, marker="o", label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
# plt.ylim(0, 1000)
# plt.xlim(2008 , 2012)
plt.legend()
```

```python
# grid
plt.plot(df['index'], df['V Kohli'], c="blue", linewidth=2, linestyle="solid", marker="o", label="Virat Kohli")
plt.plot(df['index'], df['RG Sharma'], c = "red", linewidth=2, marker="o", label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
plt.legend()
plt.grid()
```

```python
# marker and markersize
plt.plot(df['index'], df['V Kohli'], c="blue", linewidth=2, linestyle="solid", marker="d", label="Virat Kohli", markersize= 10)
plt.plot(df['index'], df['RG Sharma'], c = "red", linewidth=2, marker="o", label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
plt.legend()
```

```python
# show
plt.plot(df['index'], df['V Kohli'], c="blue", linewidth=2, linestyle="solid", marker="d", label="Virat Kohli", markersize= 10)
plt.plot(df['index'], df['RG Sharma'], c = "red", linewidth=2, marker="o", label="Rohit Sharma")
plt.xlabel("Year")
plt.ylabel('Total Runs')
plt.title('Virat Kohali and Rohit Sharma Runs Overview')
plt.legend()
plt.show()
```

# Pie Chart
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
a = [50,30,20,60,100]
b = ['abhay', 'vijay','raju', 'rahul', 'rohit']
plt.pie(a)
plt.show()
```

```python
# dataset (gayle-175.csv)
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'])
plt.show()
```

```python
# use autopct for % 
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.2f%%')
plt.show()
```

```python
# use Colors
plt.pie(df['batsman_runs'], labels=df['batsman'], colors=['red', 'green', 'pink', 'yellow', 'gold'])
plt.show()
```

```python
# explode & shadow
plt.pie(df['batsman_runs'], labels=df['batsman'], explode=[0.3,0,0, 0, 0, 0], shadow=True)
plt.show()
```
