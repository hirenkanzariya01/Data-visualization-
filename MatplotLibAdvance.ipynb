```python
# import library 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

 # Colored Scatterplots

```python
df = pd.read_csv('Datasets/iris.csv')
df.head()
```

```python
df['Species'] = df['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, "Iris-virginica":2 })
```

```python
df.head()
```

```python
# cmap and alpha 
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c = df['Species'], alpha=1)
plt.show()
```

```python
# colorbar
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c = df['Species'], alpha=1)
plt.colorbar()
plt.show()
```
